```python
covid['confirmed']=covid['confirmed'].fillna(covid['confirmed'].mode()[0])
covid['total_dead']=covid['total_dead'].fillna(covid['total_dead'].mode()[0])
```

```python
covid.isna().sum()
```

```python
covid=covid.rename(columns={'2. high':'stock_high','3. low':'stock_low'})
```

```python
covid
```

```python
fig, (ax1,ax2) = plt.subplots(2, 1, figsize=(12, 10))
ax1.plot(covid['date'], covid['stock_high'], label='High')
ax1.plot(covid['date'], covid['stock_low'], label='Low')
ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Price')
ax1.set_title('Daily High and Low Prices')
ax1.legend()
ax2.bar(covid['date'], covid['confirmed'], label='Confirmed Cases')
ax2.bar(covid['date'], covid['total_dead'], label='Total Deaths')
ax2.set_xlabel('Date')
ax2.set_ylabel('Number of Cases/Deaths')
ax2.set_title('COVID-19 Cases and Deaths')
ax2.legend()

plt.tight_layout()
plt.show()
```

To create the stock prices plot, I would use a line plot with the date on the x-axis and the high and low prices on the y-axis. This will give a clear visual representation of the daily fluctuations in the stock prices.

To create the COVID-19 cases and deaths plot, I would use two bar plots side by side with the date on the x-axis and the number of cases and deaths on the y-axis. This will provide an overview of the progression of the pandemic over time.
